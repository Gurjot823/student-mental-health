```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Student Mental health.csv")
df.head()
```

```python
df.rename(columns = {
    "Choose your gender" : "Gender",
    "What is your course?": "Course",
    "Your current year of Study": "Year",
    "What is your CGPA?": "CGPA",
    "Marital status": "MS",
    "Do you have Depression?": "Depression",
    "Do you have Anxiety?": "Anxiety",
    "Do you have Panic attack?": "Panic",
    "Did you seek any specialist for a treatment?": "Specialist",
}, inplace = True)

df.head()
```

```python
df.Gender.value_counts()
```

```python
df.Gender.replace({"Female": 0, "Male": 1}, inplace = True)
df.head()
```

```python
df.Age.value_counts()
```

```python
df = pd.get_dummies(df, drop_first = True, columns=["Course"])
df.head()
```

```python
df.Year.value_counts()
```

```python
df["Year"] = df["Year"].str.title()
df.Year.value_counts()
```

```python
df.Year.replace({"Year 1" : 1, "Year 2" : 2, "Year 3" : 3, "Year 4" : 4}, inplace = True)
df.head()
```

```python
set(df.CGPA.values)
```
